###**Representation Of the Cause Category in Pie Chart**

### Here the pie chart depicts that which cause constitutes how much of the total accidents caused





```python
plt.figure(figsize=(10, 20))
df['Cause category'].value_counts().plot.pie( autopct='%1.1f%%',radius=0.8,explode=[0.08,0,0,0,0,0],shadow=True)
plt.title('Percentage Distribution of Cause Categories')
plt.ylabel('')
plt.legend(loc='lower left',title='Cause Category')
plt.show()
```

#How much each Cause sub category constitutes in the total number of accidents caused in the country

```python
plt.figure(figsize=(40, 20))
df['Cause Subcategory'].value_counts().plot.pie( autopct='%1.1f%%',radius=1.0)
plt.title('Percentage Distribution of Cause Subcategories')
plt.ylabel('')
plt.legend(title='Cause Category')
plt.show()
```

### **Top Five Cause Subcategories Within Each Cause Category**

It Gives the idea of the top 5 sub causes in the cause category for the total road accidents caused .

```python
top_subcategories = df.groupby(['Cause category', 'Cause Subcategory'])['Count'].sum().reset_index()
top_subcategories = top_subcategories.sort_values(by='Count', ascending=False).groupby('Cause category').head(5)
top_subcategories.pivot(index='Cause Subcategory', columns='Cause category', values='Count').plot(kind='barh')
plt.title('Top Five Cause Subcategories Within Each Cause Category')
plt.ylabel('Cause Subcategory')
plt.xlabel('Number of Accidents')
plt.show()
```

# Violen Plot usage

### The violin plot is used to visualize the distribution of the 'Count' variable across different 'Cause category' groups. It allows you to compare the shapes of these distributions and identify any differences.

In the below graph ,we used voilen plot becuase , it helps depict that there a lot of values in the region of 0- 500 .So we can see that the width of the violen plot is changing for each cause category

```python
figsize = (12, 1.2 * len(df['Cause category'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(df, x='Count', y='Cause category', inner='box', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```

## By using Plotly library we can  labels the bars in the graph itself and specify the figure size in the plotly by using update_layout in which we can specify the width & height of the figure

# In each city how many are killed and the total no of accidents caused in that city.Using plotly

We use this cell because for the next plot we change the backgroud to dark

```python
pio.templates.default = "plotly_dark"
```

```python
df3=df.groupby(['Million Plus Cities','Outcome of Incident'])['Count'].sum().reset_index()
df3.sort_values(by='Count', ascending=False).groupby('Outcome of Incident')
df4=df3.drop(df3[df3['Outcome of Incident']!='Persons Killed'].index)
df4= df4.sort_values('Outcome of Incident', ascending=False)
fig1 = px.bar(df4, x='Count', y='Million Plus Cities', orientation='h', text='Count',color='Count', labels={'Count': 'People Injured'},color_continuous_scale='sunsetdark')
fig1.update_layout(width=1000,height=1000,title='Number of People Injured in Each City', xaxis_title='Number of People Injured')
fig1.show()
```

# In each city ,how many people are injured and their corresponding number of accidents.

```python
pio.templates.default = "plotly_white"
```

```python
df3=df.groupby(['Million Plus Cities','Outcome of Incident'])['Count'].sum().reset_index()
df3.sort_values(by='Count', ascending=False).groupby('Outcome of Incident')
df5=df3.drop(df3[df3['Outcome of Incident']!='Total Injured'].index)
df5 = df5.sort_values('Outcome of Incident', ascending=False)
fig1 = px.bar(df5, x='Count', y='Million Plus Cities', orientation='h', text='Count',color='Count', labels={'Count': 'People Killed'})
fig1.update_layout(width=1000,height=1000,title='Number of People Killed in Each City', xaxis_title='Number of People Killed')
fig1.show()
```
